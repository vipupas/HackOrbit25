```python
from ultralytics import YOLO
import cv2

# Load pre-trained YOLO model (trained on COCO)
model = YOLO("yolov8n.pt")  # or yolov8s.pt for better accuracy

# Open video
cap = cv2.VideoCapture(r"C:\Users\ravis\Videos\VIdeo\ved\Converted_folder\100 futa pillibhit fix 1 morning.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Run YOLO detection
    results = model(frame)

    # Visualize results
    annotated_frame = results[0].plot()  # draws boxes and labels

    # Show frame
    cv2.imshow("YOLO Vehicle Detection", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
